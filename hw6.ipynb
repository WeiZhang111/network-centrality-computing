Laplacian centrality:

Laplacian centrality is computed based on the definition on the paper http://www.math.wvu.edu/~cqzhang/Publication-files/my-paper/INS-2012-Laplacian-W.pdf

```python
def lap_cent(graph):
    vs = graph.nodes()
    degrees = graph.degree(weight=None)
    den = sum(v**2 + v for i,v in degrees)
    result = []
    for v in vs:
        neis = graph.neighbors(v)
        loc = degrees[v]
        nei = 2*sum(degrees[i] for i in neis)
        val = (loc**2 + loc + nei)/den
        result.append(val)
    return result
```

```python
laplacian_centrality_1 = lap_cent(G1)
# create a dictionary and sort 
lap_cetr_dics_1 = {}
keys = list(range(0,50))
for i in keys:

    lap_cetr_dics_1[i] = laplacian_centrality_1[i]
{k: v for k, v in sorted(lap_cetr_dics_1.items(), key=lambda item: item[1])}
```

### Discussion:



(1) Eigenvector centrality ranks nodes based on connections to important nodes. The ranking of a node is not just related to its own connections but also related to the connections of its neighbors.
(2) Pagerank centrality differs from eigenvector centrality by considering link directions. 
(3) HITS centrality (Hyperlink-Induced Topic Search centrality) ranks nodes in a network by assigning two scores to the nodes, i.e. authority and hub value. They reflect the value of content and value of its links to other pages, respectively. Therefore, HITS centrality effectively considers the pointing out and pointing in edges of a node. 
(4) Degree centrality ranks a set of nodes by considering the numbre of links that a node has.
(5) Betweenness centrality is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. Therefore, from above plotted figures, top nodes with highest betweenness centrality have obvious "bridging" functions.
(6) In Laplacian centrality, the importance (centrality) of a vertex v is reflected by the drop of the Laplacian energy of the network to respond to the deactivation (deletion) of the vertex from the network. Laplacian centrality unveils more structural information about connectivity and density around the node being ranked. Comparing with other standard centrality measures proposed for weighted networks (e.g. degree, closeness or betweenness centrality), Laplacian centrality is an intermediate measuring between global and local characterization of the importance (centrality) of a vertex.
(7) Closeness centrality represents the average distance, or average shortest path, to all other vertices in the network. It reflects the centrality more globally. Hence, the top nodes with highest closeness centrality are within the center of the entire network.
